```python
from openai import OpenAI
client = OpenAI()

def ask_llm(prompt: str, model: str = "gpt-4o-mini"):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )
    return response.choices[0].message.content

# Test v·ªõi c√¢u h·ªèi v·ªÅ t√†i ch√≠nh
result = ask_llm("T√¥i chi 5 tri·ªáu ƒë·ªìng/th√°ng cho ƒÉn u·ªëng. ƒêi·ªÅu n√†y c√≥ h·ª£p l√Ω kh√¥ng?")
print(result)
```

```python
# Augmented LLM: Multi-step Reasoning
def analyze_multistep(expense: str, income: int):
    # Step 1: Extract info
    extract = ask_llm(f"Extract t·ª´ '{expense}' ra JSON: item, total_cost, monthly_payment")
    
    # Step 2: Analyze
    analysis = ask_llm(f"Ph√¢n t√≠ch: {extract}, Thu nh·∫≠p: {income}. Risk level?")
    
    # Step 3: Recommend
    advice = ask_llm(f"D·ª±a tr√™n {analysis}, n√™n mua kh√¥ng? ƒê·ªÅ xu·∫•t?")
    
    return advice

# Test
analyze_multistep("iPhone 15 gi√° 24tr, tr·∫£ g√≥p 12 th√°ng", 15_000_000)
```

```python
# N√ÇNG C·∫§P #2: Self-Improvement qua Reflection
def analyze_with_reflection(question: str, max_rounds: int = 2):
    answer = ask_llm(question)
    
    for round in range(max_rounds):
        print(f"\n Round {round+1}")
        print(f"Answer: {answer[:100]}...")
        
        # Judge
        judge = ask_llm(f"ƒê√°nh gi√° '{answer}'. Thi·∫øu g√¨? N·∫øu OK ‚Üí 'APPROVED'")
        print(f"Judge: {judge[:100]}...")
        
        if "APPROVED" in judge.upper():
            break
        
        # Improve
        answer = ask_llm(f"S·ª≠a '{answer}' d·ª±a tr√™n: {judge}")
    
    return answer

# Test
analyze_with_reflection("T√¥i 25 tu·ªïi, l∆∞∆°ng 15tr. N√™n ti·∫øt ki·ªám th·∫ø n√†o?")
```

# 2. Multi-agent systems (No Tools)

```python
# Multiple Models - Different Perspectives
def multi_agent_analysis(question: str):
    agents = {
        "Practical": ("gpt-4o-mini", "practical advisor"),
        "Strategic": ("gpt-4.1-mini", "long-term strategist"),
        "Risk": ("gpt-4.1-mini", "risk manager")
    }
    
    answers = {}
    for name, (model, role) in agents.items():
        prompt = f"As {role}, analyze: {question}"
        answers[name] = ask_llm(prompt, model)
    
    return answers

# Test
answers = multi_agent_analysis("28 tu·ªïi, 25tr/th√°ng. Thu√™ nh√† hay mua nh√†?")
for name, ans in answers.items():
    print(f"\n{name}: {ans[:150]}...")
```

```python
# evaluate agent (LLM-as-a-evaluator) - judge pattern + structure output
class JudgeDecision(BaseModel):
    winner: str
    reasoning: str
    final_advice: str

def judge_answers(question: str, answers: dict):
    formatted = "\n".join([f"{k}: {v}" for k, v in answers.items()])
    
    prompt = f"""
    Question: {question}
    Answers: {formatted}
    
    Ch·ªçn best answer, explain why, give final advice.
    JSON format: {{"winner": "", "reasoning": "", "final_advice": ""}}
    """
    
    completion = client.beta.chat.completions.parse(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        response_format=JudgeDecision
    )
    
    return completion.choices[0].message.parsed

# Test
decision = judge_answers("Thu√™ hay mua nh√†?", answers)
print(f"Winner: {decision.winner}")
print(f"Advice: {decision.final_advice}")
```

### Budget Analyzer Tool

```python
def calculate_budget_status(category: str, year: int, month: int):
    data = df[
        (df["expense_category"] == category) &
        (df["year"] == year) &
        (df["month"] == month)
    ]
    
    spent = int(data["amount"].sum())
    budget = BUDGETS.get(category, 0)
    remaining = budget - spent
    pct = round(spent / budget * 100, 1) if budget > 0 else 0
    status = "OVER" if remaining < 0 else "WARNING" if pct > 80 else "OK"

    return {
        "category": category,
        "spent": spent,
        "budget": budget,
        "remaining": remaining,
        "percentage": pct,
        "status": status
    }
# test
calculate_budget_status("food", 2024, 11)
```

```python
# get top expense (KRW)
def get_top_expenses(category: str, year: int, month: int, top_n: int = 5):
    top = df[
        (df["expense_category"] == category) &
        (df["year"] == year) &
        (df["month"] == month)
    ].nlargest(top_n, "amount")

    return top[["date", "name", "amount"]].to_dict("records")

get_top_expenses("food", 2024, 11)
```

```python
# Email notification 
import os
import requests

RESEND_API_KEY = os.getenv("RESEND_API_KEY")
EMAIL_FROM = os.getenv("EMAIL_FROM", "onboarding@resend.dev")
EMAIL_TO = os.getenv("EMAIL_TO")

def send_email(subject: str, message: str, priority: str = "normal"):
    if not RESEND_API_KEY or not EMAIL_TO:
        return {"status": "skipped", "reason": "resend not configured"}

    priority_subjects = {
        "low": "Finance Info",
        "normal": "Monthly Update",
        "high": "Budget Warning",
        "urgent": "Budget Exceeded"
    }

    payload = {
        "from": EMAIL_FROM,
        "to": [EMAIL_TO],
        "subject": priority_subjects.get(priority, subject),
        "html": f"<p>{message}</p>"
    }

    headers = {
        "Authorization": f"Bearer {RESEND_API_KEY}",
        "Content-Type": "application/json"
    }

    response = requests.post(
        "https://api.resend.com/emails",
        json=payload,
        headers=headers
    )

    if response.status_code >= 400:
        # print("Resend status:", response.status_code)
        # print("Resend body:", response.text)
        return {"status": "error", "error": response.text}

    return {"status": "sent", "to": EMAIL_TO}
```

## Finance Agent (real data, deterministic)

```python
import json

def finance_agent(question: str, max_iter: int = 8):
    messages = [{
        "role": "system",
        "content": "You are a personal finance assistant. Use tools only with real data."
    }]
    messages.append({"role": "user", "content": question})

    for _ in range(max_iter):
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            tools=tools
        )

        choice = response.choices[0]

        if choice.finish_reason != "tool_calls":
            return choice.message.content

        messages.append(choice.message)

        for call in choice.message.tool_calls:
            name = call.function.name
            args = json.loads(call.function.arguments)

            print(f" {name}({args})")
            result = TOOL_MAP[name](**args)

            messages.append({
                "role": "tool",
                "tool_call_id": call.id,
                "content": json.dumps(result, ensure_ascii=False)
            })

    return "Agent stopped (max iterations)"
```

## Autonomous Agent

```python
# Evaluator (finance-aware)
from pydantic import BaseModel

class Evaluation(BaseModel):
    is_acceptable: bool
    has_numbers: bool
    has_actions: bool
    feedback: str


def evaluate_advice(question: str, advice: str):
    prompt = f"""
Q: {question}
A: {advice}

Evaluate for personal finance advice.
Return JSON with:
- is_acceptable
- has_numbers (‚Ç©, %, months)
- has_actions (clear next steps)
- feedback
"""

    completion = client.beta.chat.completions.parse(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        response_format=Evaluation
    )

    return completion.choices[0].message.parsed
```

```python
# retry loop (no hallucination)
def finance_agent_retry(question: str, max_attempts: int = 3):
    for attempt in range(max_attempts):
        print(f"\n Attempt {attempt + 1}")

        advice = finance_agent(question)
        eval = evaluate_advice(question, advice)

        if eval.is_acceptable and eval.has_numbers and eval.has_actions:
            print(" PASSED")
            return advice

        print(f" {eval.feedback}")
        question += f"\n[IMPROVE]: {eval.feedback}"

    return " Failed after retries"
```

```python
# state management
class State:
    def __init__(self):
        self.tasks = []
        self.completed = []
    
    def add_task(self, desc: str):
        self.tasks.append({"id": len(self.tasks), "desc": desc, "done": False})
        return f"Added {len(self.tasks)-1}"
    
    def complete_task(self, task_id: int):
        if task_id < len(self.tasks):
            self.tasks[task_id]["done"] = True
            self.completed.append(self.tasks[task_id])
            return "Done"
        return "Invalid"
    
    def status(self):
        pending = [t for t in self.tasks if not t["done"]]
        return {"pending": len(pending), "completed": len(self.completed)}

state = State()

# Add task tools
def add_task(description: str):
    return state.add_task(description)

def complete_task(task_id: int):
    return state.complete_task(task_id)

def task_status():
    return state.status()
```

```python
# Autonomous Agent (real execution)
# Add task tools to schema
task_tools = [
    {"type": "function", "function": {
        "name": "add_task",
        "description": "Th√™m task v√†o plan",
        "parameters": {"type": "object", "properties": 
                      {"description": {"type": "string"}}, "required": ["description"]}
    }},
    {"type": "function", "function": {
        "name": "complete_task",
        "description": "ƒê√°nh d·∫•u task done",
        "parameters": {"type": "object", "properties": 
                      {"task_id": {"type": "integer"}}, "required": ["task_id"]}
    }}
]
all_tools = tools + task_tools
all_tool_map = {**TOOL_MAP, "add_task": add_task, "complete_task": complete_task}

def autonomous_agent(goal: str, max_iter: int = 12):
    messages = [{
        "role": "system",
        "content": """
    You are an autonomous personal finance agent.

    You already have access to:
    - Transaction data dataframe (df)
    - Budget configuration (BUDGETS)
    - Expense categories
    - Year and month information inside the data

    DO NOT ask the user for missing information.
    Your job is to:
    - Analyze all available data
    - Identify categories exceeding budget
    - Send email alerts ONLY when status is WARNING or OVER
    - Act autonomously without clarification questions
    """
    }]

    messages.append({"role": "user", "content": goal})

    for _ in range(max_iter):
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            tools=all_tools
        )

        choice = response.choices[0]

        if choice.finish_reason != "tool_calls":
            return choice.message.content

        messages.append(choice.message)

        for call in choice.message.tool_calls:
            name = call.function.name
            args = json.loads(call.function.arguments)

            print(f"üîß {name}({args})")
            result = all_tool_map[name](**args)

            messages.append({
                "role": "tool",
                "tool_call_id": call.id,
                "content": json.dumps(result, ensure_ascii=False)
            })

    return "Autonomous loop ended"
```
